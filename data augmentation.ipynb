```python
import os
import numpy as np
import soundfile as sf  # Utilisez la bibliothque soundfile pour la lecture/criture audio
import librosa

# Chemin du dossier contenant les fichiers audio originaux
dossier_audio = "/Users/ariel/Desktop/Hubert-PING/cats_dogs"

# Crer un nouveau dossier pour les fichiers audio traits
dossier_sortie = os.path.join(dossier_audio, "stretching")
os.makedirs(dossier_sortie, exist_ok=True)  # Cre le dossier s'il n'existe pas dj

# Factor de dcalage temporel en chantillons (ajustez selon vos besoins)
factor_decalage = 3000

# Parcourez tous les fichiers audio dans le dossier
for fichier_audio in os.listdir(dossier_audio):
    if fichier_audio.endswith(".wav"):  # Assurez-vous que les fichiers sont au format WAV (vous pouvez ajuster selon votre format)
        chemin_complet_entree = os.path.join(dossier_audio, fichier_audio)

        try:
            # Lisez le fichier audio original
            audio_original, taux_echantillonnage = sf.read(chemin_complet_entree)

            # Convertissez le signal audio en mono si ncessaire
            if audio_original.ndim > 1:
                audio_original = np.mean(audio_original, axis=1)
            
            # Appliquez le dcalage temporel
            rate = 0.9
            audio_decale = librosa.effects.time_stretch(audio_original, rate = rate)

            # Convertissez le rsultat en un format acceptable par soundfile.write
            audio_decale = np.asarray(audio_decale, dtype=np.float32)

            # Construisez le nouveau nom de fichier avec "stretching"
            nouveau_nom_fichier = fichier_audio.replace(".wav", "_stretching.wav")

            # Chemin complet pour le fichier de sortie dans le nouveau dossier
            chemin_complet_sortie = os.path.join(dossier_sortie, nouveau_nom_fichier)

            # crivez le fichier audio dcal dans le nouveau dossier avec le nouveau nom
            sf.write(chemin_complet_sortie, audio_decale, taux_echantillonnage)

        except Exception as e:
            print(f"Erreur lors du traitement de {fichier_audio}: {e}. Fichier ignor.")
            continue

print("Dcalage temporel termin pour tous les fichiers audio dans le dossier.")
```
